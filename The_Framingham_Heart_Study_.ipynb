```python
# Hypothesis Testing
# Chi-squared test for independence
from scipy.stats import chi2_contingency

# Existing code for the chi-squared test
contingency_table = pd.crosstab(df['cholesterol_category'], df['TenYearCHD'])
chi2, p, _, _ = chi2_contingency(contingency_table)

print(f"Chi-squared statistic: {chi2}")
print(f"P-value: {p}")
```

```python
# Conclusion and Interpretation
if p < 0.05:
    print("Reject the null hypothesis. There is a significant difference in CHD risk between high cholesterol and normal cholesterol individuals.")
    print("This suggests that high cholesterol is associated with a higher risk of getting 10-year CHD.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in CHD risk based on cholesterol levels.")
```

## **CONCLUSION**


**Chi-squared Test Result**: The chi-squared test for independence yielded a chi-squared statistic of approximately 18.81 and an extremely low p-value of approximately 0.0000145 (less than the typical significance level of 0.05).

**Conclusion**: Based on the statistical analysis and insights from data visualizations, we reject the null hypothesis. Our findings suggest a significant difference in the ten-year CHD risk between individuals with 'High' and 'Normal' cholesterol levels. High cholesterol is associated with a higher risk of developing CHD within ten years.
